```python
# Question 1
# Max from the list
def max_from_list(list_of_values: list) -> int:
  return max(list_of_values)
print(max_from_list([2, 10, 19, 4]))

# Concatenate 2 lists
def concat_2_lists(list1: list, list2: list) -> list:
  return list1 + list2
print(concat_2_lists([10 , 19, 20], [3, 4, 11]))

# reverse a list
def reverse_list(list1: list) -> list:
  list1.reverse()
  return list1
print(reverse_list([10, 18, 31, 99]))

# interchange first and last elements in the list
def interchange(list1: list) -> list:
  tmp = list1[0]
  list1[0] = list1[-1]
  list1[-1] = tmp
  return list1
print(interchange([10, 100, 11]))

def searchSubString(list1: list, string: str) -> bool:
  return any(string in string for string in list1)
print(searchSubString(['My Name', 'is', 'Rishabh'], 'Rish'))
```

```python
# Question 2
def delete_element_by_index(list1: list, idx: int) -> list:
  if idx >= len(list1):
    return 'No element exists'
  list1.pop(idx)
  return list1
print(delete_element_by_index([10, 89, 30, 55], 2))
```

```python
# Question 9
def remove_dups_from_list_and_reverse(list1: list) -> list:
  visited = set()
  for i in range(0, len(list1)):
    if list1[i] in visited:
      list1.pop(i)
    else:
      visited.add(list1[i])
  list1.reverse()
  return list1
print(remove_dups_from_list_and_reverse([10, 8, 0, 8]))
```

```python
# Question 11
def isSymmetrical(a) -> bool:

    n = len(a)
    flag = 0

    if n%2:
        mid = n//2 +1
    else:
        mid = n//2

    start1 = 0
    start2 = mid

    while(start1 < mid and start2 < n):

        if (a[start1]== a[start2]):
            start1 = start1 + 1
            start2 = start2 + 1
        else:
            flag = 1
            break

    if flag == 0:
        return True
    else:
        return False


def isPalindrome(a):

    mid = (len(a)-1)//2
    start = 0
    last = len(a)-1
    flag = 0

    while(start <= mid):

        if (a[start]== a[last]):

            start += 1
            last -= 1

        else:
            flag = 1
            break;

    if flag == 0:
        return True
    else:
        return False


print(isSymmetrical('byebye'))
print(isPalindrome('malayalam'))
```

```python
# Question 12
fib = [0, 1]

def fibonacci(n):
    if n<0:
        print("unacceptable input")
    elif n<= len(fib):
        return fib[n-1]
    else:
        temp_fib = fibonacci(n-1)+fibonacci(n-2)
        fib.append(temp_fib)
        return temp_fib


print(fibonacci(9))
```

```python
# Question 13
def lcsubstr(S, T):
  m = len(S)
  n = len(T)
  counter = [[0]*(n+1) for x in range(m+1)]
  longest = 0
  lcs_set = set()
  for i in range(m):
      for j in range(n):
          if S[i] == T[j]:
              c = counter[i][j] + 1
              counter[i+1][j+1] = c
              if c > longest:
                  lcs_set = set()
                  longest = c
                  lcs_set.add(S[i-c+1:i+1])
              elif c == longest:
                  lcs_set.add(S[i-c+1:i+1])

  return lcs_set
print(lcsubstr("There was a lock on the door", "We are currently in lockdown"))
print(lcsubstr("There is a lady on the mountain with an umbrella",
"I love to look at the mountain when it is snowcapped"))
```

```python
# Question 18
import pandas as pd
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np


train = pd.read_csv('train.csv')
# filling in missing values with 0
for col in train.columns:
  train[col] = pd.to_numeric(train[col], errors='coerce')
print(train.isnull().sum())
train = train.replace([np.nan, -np.inf], 0)

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(train.values, i) for i in range(train.shape[1])]
vif['variable'] = train.columns
```

```python
vif.sort_values(by = 'VIF', ascending = False)
```

```python
from sklearn.ensemble import RandomForestClassifier
X = train.drop('Response', axis=1)
y = train['Response']

# Encoding categorical variables
X = pd.get_dummies(X)

clf = RandomForestClassifier()
clf.fit(X, y)

feature_importances = pd.DataFrame(clf.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)

print(feature_importances.head(3))
```
