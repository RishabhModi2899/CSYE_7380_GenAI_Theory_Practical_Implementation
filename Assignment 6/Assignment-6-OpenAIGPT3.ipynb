```python
import re

def preprocess_message(message):
  # Remove special characters, numbers, and punctuation
    cleaned_message = re.sub(r'[^a-zA-Z\s]', '', message)

    # Convert to lowercase
    cleaned_message = cleaned_message.lower()


    return cleaned_message
```

```python
df['text'] = df['text'].apply(preprocess_message)
df
```

```python
def classify_emails(email):
  response = openai.Completion.create(
      engine="text-davinci-002",
      prompt=f"Classify the following message as spam or ham: '{email}'",
      max_tokens=5
    )
  return response.choices[0].text.strip()
```

```python
df_sample = df_cleaned.head(100)
df_sample
```

```python
correct_preds = 0
for idx,row in df_sample.iterrows():
  predicted_label = classify_emails(row['text'])
  print(f"predicted Label- {predicted_label}, text- {row['text']}, real label- {row['label']}")
  print('\n')
  if predicted_label == row['label']:
    correct_preds += 1

acc = correct_preds / len(df_sample) * 100
print(f"Accuracy: {acc:.2f}%")
```
