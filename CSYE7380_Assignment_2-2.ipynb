```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def text_similarity(text1, text2):
  vect = TfidfVectorizer().fit_transform([text1, text2])
  vectors = vect.toarray()
  cosine_sim = cosine_similarity(vectors[0].reshape(1, -1), vectors[1].reshape(1, -1))[0][0]
  return cosine_sim

text1 = "My name is Rishabh Modi"
text2 = "I have a friend name RIshabh MOdi"
text3 = """Machine learning is a method of data analysis that automates analytical model building."""
text4 = """Machine learning automates the building of analytical models by using algorithms that iteratively learn from data."""

sim = text_similarity(text1, text2)
print(f"The similarity between the 2 texts is : {sim:.2f}")
sim_1 = text_similarity(text3, text4)
print(f"The similarity between the 2 texts is : {sim_1:.2f}")
```

```python
'''
  Sentence Similarity using BERT (Bidirectional Encoder Representations from Transformers)
'''
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import torch

model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

def get_sentence_embedding(sentence):
  inputs = tokenizer(sentence, return_tensors='pt', max_length=128, truncation=True)
  with torch.no_grad():
    outputs = model(**inputs)

  last_layer_hidden_states = outputs.last_hidden_state

  sentence_embedding = last_layer_hidden_states.mean(dim=0)

  return sentence_embedding

def get_similarity(sentence1, sentence2):
  embed_1 = get_sentence_embedding(sentence1)
  embed_2 = get_sentence_embedding(sentence2)

  similarity = cosine_similarity(embed_1.numpy(), embed_2.numpy())

  return similarity[0][0]

sentence1 = 'I am a AI bot trying to capture the world'
sentence2 = 'A AI bot who is capable of capturing the world'
text1 = "My name is Rishabh Modi"
text2 = "I have a friend name RIshabh MOdi"
text3 = """Machine learning is a method of data analysis that automates analytical model building."""
text4 = """Machine learning automates the building of analytical models by using algorithms that iteratively learn from data."""

print(f"similarity score: { get_similarity(sentence1, sentence2) }")
print(f"similarity score: { get_similarity(text1, text2) }")
print(f"similarity score: { get_similarity(text3, text4) }")

```
